```python
# Getting data from World Bank  API

import pandas as pd
import matplotlib.pyplot as plt
import wbgapi as wb
```

```python
wb.series.info(q = 'business')
```

```python
# 1 business density
New_Business_Den = wb.data.DataFrame('IC.BUS.NDNS.ZS', wb.region.members('EMU'), time = range(2015,2019))
New_Business_Den
```

```python
# 2 cost of startup
Cost = wb.data.DataFrame('IC.REG.COST.PC.ZS', wb.region.members('EMU'), time = range(2015,2019))
Cost
```

```python
# 3 Time required
time_required = wb.data.DataFrame('IC.REG.DURS', wb.region.members('EMU'), time = range(2015,2019))
time_required
```
